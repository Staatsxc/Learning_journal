```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data-2.csv')
```

```python
df
```

```python
# Membuat fungsi untuk check apakah pokemon memiliki satu atau dua tipe
kolom = []

def kali (a,b):
    result = a * b
    return result
```

```python
for idx in range(len(df)):
    res = kali(df.loc[idx, 'Quantity'],df.loc[idx, 'UnitPrice'])
    kolom.append(res)
```

```python
df['Revenue'] = kolom
```

```python
df_group = df.groupby(by = 'Country').sum()
```

```python
del df_group['CustomerID']
```

```python
df_group = df_group.sort_values(by = 'Revenue', ascending = False)
df_group.head(10)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.scatter(x = df_group['Quantity'], 
            y = df_group['Revenue'],
           )
```
