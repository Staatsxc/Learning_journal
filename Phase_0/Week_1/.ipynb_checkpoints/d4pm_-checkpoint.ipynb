## **Column Manipulation**

```python
nba = df.copy()
nba
```

**rename column**

```python
nba.head(2)
```

```python
nba.rename(columns = {'game_location': 'loc', 'game_result':'result'}, inplace=True)
nba.rename(columns = {'loc': 'location'}, inplace=True)
```

```python
nba['location'].head(2)
```

**drop column**

```python
nba.head(2)
```

```python
nba.drop(labels=['elo_i', 'elo_n'], axis=1, inplace=True)
```

```python
nba.head(2)
```

**creating column**

```python
nba['pts_diff'] = nba['pts'] - nba['opp_pts']
```

```python
nba.tail(5)
```

```python
range(len(nba))
```

```python
# siapkan list kosong untuk tempat penampung
# siapkan logic/prosedur looping
# hasil loop masukkan ke list kosong
# list dijadikan kolom baru

# kolom_pts = nba.iloc[:, 10]
# kolom_opp_pts = nba.iloc[:, 14]

hasil = []

for baris in range(len(nba)):
  selisih = nba.iloc[baris, 10] - nba.iloc[baris, 14]
  hasil.append(selisih)

hasil
```

```python
nba['diff_loop'] = hasil
```

```python
nba
```

```python
nba_clean = nba.dropna(axis=0)
nba_clean
```

```python
hasil = []

for baris in range(len(nba_clean)):
  selisih = nba.iloc[baris, 10] - nba.iloc[baris, 14]
  hasil.append(selisih)

hasil
```

```python
nba_clean.reset_index(inplace=True)
```

```python
list(range(len(nba_clean)))
```

```python
hasil_loc = []

for baris in range(len(nba_clean)):
    selisih = nba_clean.loc[baris, 'pts'] - nba_clean.loc[baris, 'opp_pts']
    hasil_loc.append(selisih)

nba_clean['loop_loc'] = hasil_loc
```

```python
nba_clean.head(5)
```

```python
def hasil(diff):
    if diff > 0:
        return 'Menang'
    elif diff < 0:
        return 'Kalah'
    else:
        return 'Imbang'
```

```python
keterangan = []

for idx in range(len(nba)):
    res = hasil(nba.loc[idx, 'pts_diff'])
    keterangan.append(res)

nba['status'] = keterangan
```
